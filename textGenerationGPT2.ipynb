```python
file_name = 'the_wonderful_wizard_of_oz.txt'
```

```python
gpt2.copy_file_from_gdrive(file_name)
```

```python
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset = file_name,
              model_name = '124M',
              steps = 1400,
              restore_from = 'fresh',
              run_name = 'run1',
              print_every = 10,
              sample_every = 200,
              save_every = 700,
            #   learning_rate=1e-5
            #   batch_size = 5,
            #   use_memory_saving_gradients = True,
            #   only_train_transformer_layers= True,
            #   accumulate_gradients = 1
            )
```

```python
gpt2.copy_checkpoint_to_gdrive(run_name = 'run1')
```

generating text from the above trained model

```python
text = gpt2.generate(sess, run_name = 'run1')
```
